```python
import os
import pandas as pd
import time

# Mesurer le temps de début
start_time = time.time()

# Spécifier le chemin du dossier contenant les fichiers texte
dossier_fichiers = "/content/drive/MyDrive/test_fff"

# Liste des encodages possibles
encodages = ['utf-8', 'latin-1', 'cp1252', 'ascii', 'utf-16', 'utf-32']

# Créer un dossier pour sauvegarder les fichiers Excel dans le chemin spécifié
dossier_sauvegarde = "/content/drive/MyDrive/finale"
os.makedirs(dossier_sauvegarde, exist_ok=True)

# Initialiser le compteur de fichiers
compteur_fichiers = 0

# Parcourir tous les fichiers texte dans le dossier
for fichier_texte in os.listdir(dossier_fichiers):
    chemin_fichier = os.path.join(dossier_fichiers, fichier_texte)

    for encodage in encodages:
        try:
            with open(chemin_fichier, "r", encoding=encodage) as fichier:
                contenu = fichier.read()
            print(f"Fichier '{fichier_texte}' lu avec l'encodage : {encodage}")

            # Convertir le contenu du fichier en DataFrame (en supposant qu'il a une structure tabulaire)
            df = pd.DataFrame([x.split('\t') for x in contenu.split('\n') if x])
            df["Fermenteur"] = int(df.loc[:, 0][1].split('=')[1])
            df["date_experience"] = df.loc[:, 0][3].split('=')[1]
            df["Operateur"] = df.loc[:, 0][df[df[0].str.startswith('Op')].index[0]].split('=')[1]
            df["Commentaire"] = df.loc[:, 0][df[df[0].str.startswith('Commentaire')].index[0]].split('=')[1]
            df["Souche"] = df.loc[:, 0][df[df[0].str.startswith('Souche')].index[0]].split('=')[1]
            df["Milieu"] = df.loc[:, 0][df[df[0].str.startswith('Milieu')].index[0]].split('=')[1]
            valeurs_a_chercher = ["[Données]", "[Donn�es]"]
            df = df.loc[:,][df.loc[df.loc[:, 0].isin(valeurs_a_chercher)].index[0] + 1:]

            # Obtenir les noms des 12 premières colonnes à partir de la première ligne
            nouvelles_colonnes = df.iloc[0, :13].tolist()

            # Obtenir les noms des colonnes restantes (à partir de la 13ème colonne)
            remaining_columns = df.columns[13:].tolist()

            # Assigner les nouveaux noms de colonnes au DataFrame
            df.columns = nouvelles_colonnes + remaining_columns

            # Supprimer la première ligne (qui a été utilisée comme noms de colonnes)
            df = df.iloc[1:]

            # Réinitialiser l'index
            df = df.reset_index(drop=True)

            # Chemin complet pour sauvegarder le fichier Excel
            chemin_excel = os.path.join(dossier_sauvegarde, f"{os.path.splitext(fichier_texte)[0]}.xlsx")

            # Sauvegarder le DataFrame dans un fichier Excel
            df.to_excel(chemin_excel, index=False)
            print(f"Le DataFrame a été sauvegardé sous format Excel dans '{chemin_excel}'.")

            # Incrémenter le compteur de fichiers
            compteur_fichiers += 1

            break  # Sortir de la boucle si la lecture est réussie

        except UnicodeDecodeError:
            print(f"Échec de la lecture avec l'encodage : {encodage}")
        except FileNotFoundError:
            print(f"Erreur : Le fichier '{chemin_fichier}' n'existe pas.")
            break
        except Exception as e:
            print(f"Une erreur s'est produite avec l'encodage {encodage} : {e}")

# Mesurer le temps de fin
end_time = time.time()

# Calculer le temps d'exécution
execution_time = end_time - start_time
print(f"Nombre total de fichiers traités : {compteur_fichiers}")
print(f"Temps d'exécution : {execution_time:.2f} secondes")
```
