```python
import pandas as pd

# Charger le fichier Excel
df = pd.read_excel('Classeur1.xlsx', engine='openpyxl')
```

```python
df
```

```python
resultats=[]
for code in df['Code'].unique():
    data = df[df["Code"] == code]

    filtered_df = data[(data['CO2'] >= 3) & (data['CO2'] <= 40)]

    filtered_df = filtered_df.reset_index(drop=True)
    filtered_df.index = range(0, len(filtered_df))

    result = filtered_df['V'].max()
    tempsmax=filtered_df.iloc[filtered_df['V'].idxmax()]['Temps']


    filtered_data_temps = data[data['Temps'] >= 50]


    max_co2 = filtered_data_temps['CO2'].max()


    tempsmaxco2= filtered_data_temps.loc[filtered_data_temps['CO2'].idxmax(), 'Temps']


    result_dict = {
    'Code':code,
    'Max_V':result,
    'Tempsmax':tempsmax,
    'max_co2':max_co2,
    'tempsmaxco2':tempsmaxco2
  }
    resultats.append(result_dict)


DF = pd.DataFrame(resultats)
```

```python
DF['Maxco2_85']=DF['max_co2']*0.85
DF['Maxco2_98']=DF['max_co2']*0.98
DF['Maxco2_70']=DF['max_co2']*0.7
DF['Maxco2_80']=DF['max_co2']*0.8
DF
```

```python

import statsmodels.formula.api as sm
resul=[]
for code in df['Code'].unique():
  data = df[df["Code"] == code]
  # Get the values as a single value instead of a Series
  maxco2_90_value = DF[DF["Code"] == code]["Maxco2_85"].iloc[0]
  maxco2_95_value = DF[DF["Code"] == code]["Maxco2_98"].iloc[0]
  #Now we compare using the single values instead of Series
  datafil=data[(data['CO2'] >= maxco2_90_value) & (data['CO2'] <= maxco2_95_value)]
  modele=sm.ols('V ~ Temps',data=datafil).fit()
  A0=modele.params.Intercept
  A1=modele.params.Temps

  result = {
    'code':code,
    'A0':float(A0),
    'A1':float(A1),
  }
  resul.append(result)
DF2 = pd.DataFrame(resul)
DF2
```

```python
DF2['FIN_FERMENTATION']=-DF2['A0']/DF2['A1']
DF2
```

```python
data_finale=pd.merge(DF,DF2,left_on='Code',right_on='code')
data_finale
```

```python
data_finale.to_excel('resultats.xlsx', index=False)
```
