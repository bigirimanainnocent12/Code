```python
import pandas as pd

# Charger le fichier Excel
df = pd.read_excel('Classeur1.xlsx', engine='openpyxl')
```

```python
df
```

```python
resultats=[]
for code in df['Code'].unique():
    data = df[df["Code"] == code]

    filtered_df = data[(data['CO2'] >= 3) & (data['CO2'] <= 40)]

    filtered_df = filtered_df.reset_index(drop=True)
    filtered_df.index = range(0, len(filtered_df))

    result = filtered_df['V'].max()
    tempsmax=filtered_df.iloc[filtered_df['V'].idxmax()]['Temps']


    filtered_data_temps = data[data['Temps'] >= 50]


    max_co2 = filtered_data_temps['CO2'].max()


    tempsmaxco2= filtered_data_temps.loc[filtered_data_temps['CO2'].idxmax(), 'Temps']


    result_dict = {
    'Code':code,
    'Max_V':result,
    'Tempsmax':tempsmax,
    'max_co2':max_co2,
    'tempsmaxco2':tempsmaxco2
  }
    resultats.append(result_dict)


DF = pd.DataFrame(resultats)
DF['Maxco2_95']=DF['max_co2']*0.95
DF['Maxco2_98']=DF['max_co2']*0.98
DF['Maxco2_92']=DF['max_co2']*0.92
DF['Maxco2_995']=DF['max_co2']*0.995
DF['Maxco2_90']=DF['max_co2']*0.90
tempsmaxco2=[]
for code in df['Code'].unique():
  dataa=df[df["Code"] == code]
  dataa[dataa['CO2']>=float(DF[DF["Code"] ==code]['Maxco2_95'].iloc[0])]['Temps'].min()
  dicti={"Code":code,

      'tempsmaxco2_95':dataa[dataa['CO2']>=float(DF[DF["Code"] ==code]['Maxco2_95'].iloc[0])]['Temps'].min(),
      'tempsmaxco2_98':dataa[dataa['CO2']>=float(DF[DF["Code"] ==code]['Maxco2_98'].iloc[0])]['Temps'].min(),
      'tempsmaxco2_92':dataa[dataa['CO2']>=float(DF[DF["Code"] ==code]['Maxco2_92'].iloc[0])]['Temps'].min(),
     'tempsmaxco2_995':dataa[dataa['CO2']>=float(DF[DF["Code"] ==code]['Maxco2_995'].iloc[0])]['Temps'].min(),
      'tempsmaxco2_90':dataa[dataa['CO2']>=float(DF[DF["Code"] ==code]['Maxco2_90'].iloc[0])]['Temps'].min()
         }

  tempsmaxco2.append(dicti)
DF3 = pd.DataFrame(tempsmaxco2)
DF3
```

```python
import statsmodels.formula.api as sm
resul=[]
for code in df['Code'].unique():
  data = df[df["Code"] == code]
  # Get the values as a single value instead of a Series
  maxco2_value_95 = DF[DF["Code"] == code]["Maxco2_95"].iloc[0]
  maxco2_value_995 = DF[DF["Code"] == code]["Maxco2_995"].iloc[0]
  #Now we compare using the single values instead of Series
  datafil_95_995=data[(data['CO2'] >= maxco2_value_95) & (data['CO2'] <= maxco2_value_995)]
  modele_95_995=sm.ols('V ~ Temps',data=datafil_95_995).fit()
  A0_95_995=modele_95_995.params.Intercept
  A1_95_995=modele_95_995.params.Temps


  maxco2_value_92 = DF[DF["Code"] == code]["Maxco2_92"].iloc[0]
  maxco2_value_98 = DF[DF["Code"] == code]["Maxco2_98"].iloc[0]
  #Now we compare using the single values instead of Series
  datafil_92_98=data[(data['CO2'] >= maxco2_value_92) & (data['CO2'] <= maxco2_value_98)]
  modele_92_98=sm.ols('V ~ Temps',data=datafil_92_98).fit()
  A0_92_98=modele_92_98.params.Intercept
  A1_92_98=modele_92_98.params.Temps




  maxco2_value_90 = DF[DF["Code"] == code]["Maxco2_90"].iloc[0]
  maxco2_value_98 = DF[DF["Code"] == code]["Maxco2_98"].iloc[0]
  #Now we compare using the single values instead of Series
  datafil_90_98=data[(data['CO2'] >= maxco2_value_90) & (data['CO2'] <= maxco2_value_98)]
  modele_90_98=sm.ols('V ~ Temps',data=datafil_90_98).fit()
  A0_90_98=modele_90_98.params.Intercept
  A1_90_98=modele_90_98.params.Temps





  result = {
    'code':code,
    'A0_95_995':float(A0_95_995),
    'A1_95_995':float(A1_95_995),
    'A0_92_98':float(A0_92_98),
    'A1_92_98':float(A1_92_98),
    'A0_90_98':float(A0_90_98),
    'A1_90_98':float(A1_90_98),
    'FIN_FERMENTATION_95_995':-float(A0_95_995)/float(A1_95_995),
    'FIN_FERMENTATION_92_98':-float(A0_92_98)/float(A1_92_98),
    'FIN_FERMENTATION_90_98':-float(A0_90_98)/float(A1_90_98)

  }
  resul.append(result)
DF2 = pd.DataFrame(resul)
DF2
```

# **Phase latence**

```python
lat=[]
for code in df['Code'].unique():
  data = df[df["Code"] == code]
  Latence=float(DF[DF['Code']==code]['Max_V'].iloc[0])
  dictio={'Code':code,
       'MAX_V_70':Latence*0.9}
  lat.append(dictio)
DF4=pd.DataFrame(lat)

tempslat=[]
for code in df['Code'].unique():
  dataaa=df[df["Code"] == code]

  dicti={"Code":code,

      'TempsMax_V_70': dataaa[dataaa['CO2']>=float(DF4[DF4["Code"] ==code]['MAX_V_70'].iloc[0])]['Temps'].min()

         }

  tempslat.append(dicti)

DF5 = pd.DataFrame(tempslat)
DF5['diff10']=DF5['TempsMax_V_70']-9






import statsmodels.formula.api as sm
resu=[]
for code in df['Code'].unique():
  data = df[df["Code"] == code]

  maxco2_value_upper = DF5[DF5["Code"] == code]["TempsMax_V_70"].iloc[0]
  maxco2_value = DF5[DF5["Code"] == code]["diff10"].iloc[0]


  datafi=data[(data['Temps'] >= maxco2_value) & (data['Temps'] <= maxco2_value_upper)]
  if datafi.empty:
    print(f"Warning: datafi is empty for code {code}. Skipping this code.")
    continue

  modele=sm.ols('V ~ Temps',data=datafi).fit()
  A0=modele.params.Intercept
  A1=modele.params.Temps


  if A1 == 0:
    print(f"Warning: A1 is zero for code {code}. Setting FIN_LATENCE to NaN.")
    FIN_LATENCE = float('nan')
  else:
    FIN_LATENCE = -float(A0)/float(A1)

  res= {
    'Code':code,
    'A0':float(A0),
    'A1':float(A1),
    'FIN_LATENCE':FIN_LATENCE
  }
  resu.append(res)
DF6= pd.DataFrame(resu)
```

```python
data_= pd.merge(DF5, DF6, left_on='Code', right_on='Code')
data_
```

```python
data_fi = pd.merge(DF, DF3, left_on='Code', right_on='Code')
data_final=pd.merge(data_fi, DF2, left_on='Code', right_on='code')
data_final.drop('code', axis=1, inplace=True)
data_finale=pd.merge(data_, data_final, left_on='Code', right_on='Code')
data_finale
```
