```python
import pandas as pd

# Charger le fichier Excel
df = pd.read_excel('Classeur1.xlsx', engine='openpyxl')
```

```python
df
```

```python
resultats=[]
for code in df['Code'].unique():
    data = df[df["Code"] == code]

    filtered_df = data[(data['CO2'] >= 3) & (data['CO2'] <= 40)]

    filtered_df = filtered_df.reset_index(drop=True)
    filtered_df.index = range(0, len(filtered_df))

    result = filtered_df['V'].max()
    tempsmax=filtered_df.iloc[filtered_df['V'].idxmax()]['Temps']


    filtered_data_temps = data[data['Temps'] >= 50]


    max_co2 = filtered_data_temps['CO2'].max()


    tempsmaxco2= filtered_data_temps.loc[filtered_data_temps['CO2'].idxmax(), 'Temps']


    result_dict = {
    'Code':code,
    'Max_V':result,
    'Tempsmax':tempsmax,
    'max_co2':max_co2,
    'tempsmaxco2':tempsmaxco2
  }
    resultats.append(result_dict)


DF = pd.DataFrame(resultats)
```

```python
DF['Maxco2_95']=DF['max_co2']*0.95
DF['Maxco2_98']=DF['max_co2']*0.98
DF['Maxco2_92']=DF['max_co2']*0.92
DF['Maxco2_995']=DF['max_co2']*0.995
DF['Maxco2_90']=DF['max_co2']*0.90
DF
```

```python

import statsmodels.formula.api as sm
resul=[]
for code in df['Code'].unique():
  data = df[df["Code"] == code]
  # Get the values as a single value instead of a Series
  maxco2_value_95 = DF[DF["Code"] == code]["Maxco2_95"].iloc[0]
  maxco2_value_995 = DF[DF["Code"] == code]["Maxco2_995"].iloc[0]
  #Now we compare using the single values instead of Series
  datafil_95_995=data[(data['CO2'] >= maxco2_value_95) & (data['CO2'] <= maxco2_value_995)]
  modele_95_995=sm.ols('V ~ Temps',data=datafil_95_995).fit()
  A0_95_995=modele_95_995.params.Intercept
  A1_95_995=modele_95_995.params.Temps


  maxco2_value_92 = DF[DF["Code"] == code]["Maxco2_92"].iloc[0]
  maxco2_value_98 = DF[DF["Code"] == code]["Maxco2_98"].iloc[0]
  #Now we compare using the single values instead of Series
  datafil_92_98=data[(data['CO2'] >= maxco2_value_92) & (data['CO2'] <= maxco2_value_98)]
  modele_92_98=sm.ols('V ~ Temps',data=datafil_92_98).fit()
  A0_92_98=modele_92_98.params.Intercept
  A1_92_98=modele_92_98.params.Temps




  maxco2_value_90 = DF[DF["Code"] == code]["Maxco2_90"].iloc[0]
  maxco2_value_98 = DF[DF["Code"] == code]["Maxco2_98"].iloc[0]
  #Now we compare using the single values instead of Series
  datafil_90_98=data[(data['CO2'] >= maxco2_value_90) & (data['CO2'] <= maxco2_value_98)]
  modele_90_98=sm.ols('V ~ Temps',data=datafil_90_98).fit()
  A0_90_98=modele_90_98.params.Intercept
  A1_90_98=modele_90_98.params.Temps





  result = {
    'code':code,
    'A0_95_995':float(A0_95_995),
    'A1_95_995':float(A1_95_995),
    'A0_92_98':float(A0_92_98),
    'A1_92_98':float(A1_92_98),
    'A0_90_98':float(A0_90_98),
    'A1_90_98':float(A1_90_98),
    'FIN_FERMENTATION_95_995':-float(A0_95_995)/float(A1_95_995),
    'FIN_FERMENTATION_92_98':-float(A0_92_98)/float(A1_92_98),
    'FIN_FERMENTATION_90_98':-float(A0_90_98)/float(A1_90_98)

  }
  resul.append(result)
DF2 = pd.DataFrame(resul)
DF2
```
