```python
import pandas as pd

# Charger le fichier Excel
df = pd.read_excel('Classeur1.xlsx', engine='openpyxl')
```

```python
df
```

```python
resultats=[]
for code in df['Code'].unique():
    data = df[df["Code"] == code]

    filtered_df = data[(data['CO2'] >= 3) & (data['CO2'] <= 40)]

    filtered_df = filtered_df.reset_index(drop=True)
    filtered_df.index = range(0, len(filtered_df))

    result = filtered_df['V'].max()
    tempsmax=filtered_df.iloc[filtered_df['V'].idxmax()]['Temps']


    filtered_data_temps = data[data['Temps'] >= 50]


    max_co2 = filtered_data_temps['CO2'].max()


    tempsmaxco2= filtered_data_temps.loc[filtered_data_temps['CO2'].idxmax(), 'Temps']


    result_dict = {
    'Code':code,
    'Max_V':result,
    'Tempsmax':tempsmax,
    'max_co2':max_co2,
    'tempsmaxco2':tempsmaxco2
  }
    resultats.append(result_dict)


DF = pd.DataFrame(resultats)
```

```python
DF['Maxco2_85']=DF['max_co2']*0.85
DF['Maxco2_98']=DF['max_co2']*0.97
DF['Maxco2_5']=DF['max_co2']*0.5
DF['Maxco2_90']=DF['max_co2']*0.9
DF
```

```python

import statsmodels.formula.api as sm
resul=[]
for code in df['Code'].unique():
  data = df[df["Code"] == code]
  # Get the values as a single value instead of a Series
  maxco2_90_value_85 = DF[DF["Code"] == code]["Maxco2_90"].iloc[0]
  maxco2_95_value_98 = DF[DF["Code"] == code]["Maxco2_98"].iloc[0]
  #Now we compare using the single values instead of Series
  datafil_85_98=data[(data['CO2'] >= maxco2_90_value_85) & (data['CO2'] <= maxco2_95_value_98)]
  modele_85_98=sm.ols('V ~ Temps',data=datafil_85_98).fit()
  A0_85_98=modele_85_98.params.Intercept
  A1_85_98=modele_85_98.params.Temps

  maxco2_90_value_5 = DF[DF["Code"] == code]["Maxco2_5"].iloc[0]
  maxco2_95_value_90 = DF[DF["Code"] == code]["Maxco2_90"].iloc[0]
  #Now we compare using the single values instead of Series
  datafil_5_90=data[(data['CO2'] >= maxco2_90_value_5) & (data['CO2'] <= maxco2_95_value_90)]
  modele_5_90=sm.ols('V ~ Temps',data=datafil_5_90).fit()
  A0_5_90=modele_5_90.params.Intercept
  A1_5_90=modele_5_90.params.Temps

  result = {
    'code':code,
    'A0_85_95':float(A0_85_98),
    'A1_85_95':float(A1_85_98),
     'A0_5_90':float(A0_5_90),
    'A1_5_90':float(A1_5_90),
    'FIN_FERMENTATION_85_98':-float(A0_85_98)/float(A1_85_98),
    'FIN_FERMENTATION_5_90':-float(A0_5_90)/float(A1_5_90),
    'fin':data["Temps"].iloc[-1]
  }
  resul.append(result)
DF2 = pd.DataFrame(resul)
DF2
```
